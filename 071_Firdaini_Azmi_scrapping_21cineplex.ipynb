```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import requests
page = requests.get('https://21cineplex.com/')
bs = BeautifulSoup(page.text, 'html.parser')
bs
```

```python
bs.find_all('div', class_='movie-desc')
```

```python
judul=[]
movie=bs.find_all('div', class_='movie-desc')
for m in movie:
  judul.append(m.find('h4').text)
print(judul)
```

```python
genre=[]
for m in movie:
  genre_film=m.find_previous('div',class_='movie')
  if genre_film:
    page2=genre_film.find('a')['href']
    new_page=requests.get(page2)
    bs2=BeautifulSoup(new_page.text, 'html.parser')
    g=bs2.find('li', class_='movie_genre')
    if g :
      for span in g.find_all('span') :
        span.decompose()
      genre.append(g.text.strip())
print(genre)
```

```python
tabel=pd.DataFrame({
    'Judul':judul,
    'Genre':genre,
})
tabel.head(100)
```
